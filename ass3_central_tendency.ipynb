```python
print(f"Original DataFrame size: {len(df)} rows")
```

```python
def remove_outliers_iqr(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    filtered_df = dataframe[(dataframe[column] >= lower_bound) & (dataframe[column] <= upper_bound)]
    return filtered_df

score_columns = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']

for column in score_columns:
    n_df = remove_outliers_iqr(df, column)

n_df
print(f"after DataFrame size: {len(df)} rows")

# their are no outliers in the given datat
```
